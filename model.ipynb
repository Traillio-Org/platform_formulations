## Codeforces

```python
df = pd.read_csv("./data/processed/after_survey/codeforces.csv")
df.head()
```

```python
X = df.iloc[:, :-1]
y = df.iloc[:, -1]
```

```python
def get_min_max(X):
    res = {
        "min" : [],
        "max" : []
    }
    for col in X.columns:
        res["min"].append(X[col].abs().min().item())
        res["max"].append(X[col].abs().max().item())
    return res
```

```python
get_min_max(X)
```

```python
def scale_function(X, min_max):
    X = X.copy()
    for i, col in enumerate(X.columns):
        X.loc[:, col] = X.loc[:, col].astype(float)
        X.loc[:, col] = (X[col] - min_max["min"][i])/(min_max["max"][i] - min_max["min"][i])
    return X
```

```python
from sklearn.preprocessing import minmax_scale
X = scale_function(X, get_min_max(X))
X.head()
```

```python
from sklearn.linear_model import LinearRegression
linear_regression = LinearRegression()
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, train_size=0.7)
```

```python
linear_regression.fit(X_train, y_train)
```

```python
linear_regression.score(X_test, y_test)
```

```python
import pickle
with open("./models/codeforces.pkl", "wb") as f:
    pickle.dump(linear_regression, f)
```

```python
linear_regression.intercept_
```
