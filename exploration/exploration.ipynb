```python
import numpy as np
import pandas as pd 
import matplotlib.pyplot as plt
import seaborn as sns
```

## LeetCode Data

```python
df = pd.read_csv("../data/raw/leetcode.csv")
df.head()
```

```python
df = df.dropna()
df.head()
```

```python
df.to_csv("../data/processed/before_survey/leetcode.csv", index=False)
df = pd.read_csv("../data/processed/after_survey/leetcode.csv")
df.head()
```

```python
corr_matrix = df.corr(numeric_only=True)
sns.heatmap(corr_matrix, annot=True)
```
